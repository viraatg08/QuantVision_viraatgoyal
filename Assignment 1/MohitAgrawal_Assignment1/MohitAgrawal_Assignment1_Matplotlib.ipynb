## Exercise 2
**Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

fig, ax = plt.subplots()

axins = ax.inset_axes([0.2,0.5,0.2,0.2])
axins.plot()

x1, x2 = 0, 1
y1, y2 = 0, 1
axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)

axins.xaxis.get_major_locator().set_params(nbins=5)
axins.yaxis.get_major_locator().set_params(nbins=5)

ax.indicate_inset_zoom(axins)

plt.tight_layout()
plt.show()
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,100)
y=2*x
z=x**2
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

fig, ax = plt.subplots()

axins = ax.inset_axes([0.2,0.5,0.4,0.4])
axins.set_title('zoom')
axins.set_xlabel('X')
axins.set_ylabel('Y')
axins.plot(x,y,c='blue')

x1, x2 = 20, 22
y1, y2 = 30, 50
axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)

plt.xlabel('X')
plt.ylabel('Z')

plt.tight_layout()
plt.plot(x,z,c='blue')
plt.show()
```
