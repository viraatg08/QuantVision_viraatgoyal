## Exercise 1

**Follow along with these steps**
* Create a figure object called fig using plt.figure()
* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
* Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
fig = plt.figure()
# x = np.arange(0,1) # Removed as per exercise requirements to use global x and y
# y=x # Removed as per exercise requirements to use global x and y
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, lw='10')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Title')
plt.show()
```

## Exercise 3

**Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax1.plot(x,z)
ax2.plot(x,y)
fig.show()
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
fig = plt.figure()
x = np.arange(0,100)
y = x
z = x**2
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax1.plot(x,z, lw=3)
ax2.plot(x,y, lw=2)
fig.show()
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
plt.subplots(nrows=1, ncols=2)
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,100)
y = x*2

plt.subplot(1, 2, 1)
plt.plot(x,y, ls = '--', lw = 2, c = 'blue')

x = np.arange(0,100)
y = x**2

plt.subplot(1, 2, 2)
plt.plot(x,y, lw = '2', c = 'y')

plt.show()
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize = (12,2))
x = np.arange(0,100)
y = x*2

plt.subplot(1, 2, 1)
plt.plot(x,y, lw = 2, c = 'blue')

x = np.arange(0,100)
y = x**2

plt.subplot(1, 2, 2)
plt.plot(x,y, ls = '--', lw = 2, c = 'r')

plt.show()
```
