**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 101)
x1 = np.arange(20, 23)
z = x**2
y = 2.5*x1 -10

fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax1.plot(x,z, color = 'blue')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
ax2.plot(x1,y, color = 'blue')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

ax2.set_title('zoom')

plt.show()
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python

x = np.arange(0, 101)
y = 2 * x
z = x**2

fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, ls = '--', color = 'blue', linewidth=1.5)

axes[1].plot(x, z, color = 'red', linewidth=1.5)

plt.show()
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
x = np.arange(0, 101)
y = 2 * x
z = x**2

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 2))

axes[0].plot(x, y, color = 'blue', linewidth=2)


axes[1].plot(x, z, ls = '--', color = 'red' ,linewidth=3)

plt.show()
```
