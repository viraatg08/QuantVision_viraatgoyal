# NumPy Exercises

Now that we've learned about NumPy let's test your knowledge. We'll start off with a few simple tasks, and then you'll be asked some more complicated questions.

#### Import NumPy as np

```python
import numpy as np
```

#### Create an array of 10 zeros

```python
arr = np.zeros(10)
print(arr)
```

#### Create an array of 10 ones

```python
print(np.ones(10))
```

#### Create an array of 10 fives

```python
print(np.full(10,5))
```

#### Create an array of the integers from 10 to 50

```python
i = 0
a = []
for i in range(0,41):
  a.append(i+10)
  i+=1
print(np.array(a))
```

#### Create an array of all the even integers from 10 to 50

```python
i = 0
a = []
for i in range(0,21):
  a.append(2*i+10)
  i+=1
print(np.array(a))
```

#### Create a 3x3 matrix with values ranging from 0 to 8

```python
i =0
a = []
for i in range(0,9):
  a.append(i)
  i+=1
b = np.array(a)
b.reshape(3,3)
```

#### Create a 3x3 identity matrix

```python
print(np.identity(3))
```

#### Use NumPy to generate a random number between 0 and 1

```python
np.random.rand(1)
```

#### Use NumPy to generate an array of 25 random numbers sampled from a standard normal distribution

```python
a = np.random.rand(25)
b = a*2
print(b)
```

#### Create the following matrix:

```python
i = 0
a = []
for i in range(0,100):
  a.append((i+1)/100)
  i+=1
b = np.array(a)
b = b.reshape(10,10)
print(b)
```

#### Create an array of 20 linearly spaced points between 0 and 1:

```python
a = []
i = 0
for i in range(0,20):
  a.append(1/(1+i))
  i = i+1
a.reverse()
b = np.array(a)
print(b)
```

## Numpy Indexing and Selection

Now you will be given a few matrices, and be asked to replicate the resulting matrix outputs:

```python
i = 0
a = []
for i in range(0,25):
  a.append(i+1)
  i+=1
b = np.array(a)
b = b.reshape(5,5)
```

```python
i = 0
a = []
for i in range(0,14):
  if((i+11)%5!=0):
    a.append(i+12)
    i+=1
b = np.array(a)
b = b.reshape(3,4)
print(b)
```

```python
print(b[1,3])
```

```python
a = b[:,0] - 10
a = a.reshape(3,1)
print(a)
```

```python
i = 0
a = []
for i in range(0,25):
  a.append(i+1)
  i+=1
b = np.array(a)
b = b.reshape(5,5)
print(b[4,:])
```

```python
i = 0
a = []
for i in range(0,25):
  a.append(i+1)
  i+=1
b = np.array(a)
b = b.reshape(5,5)
print(b[3:,:])
```
