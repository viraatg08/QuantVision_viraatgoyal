## Exercise 1

**Follow along with these steps**
* Create a figure object called fig using plt.figure()
* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
* Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
plt.title('title')
plt.xticks([0,20,40,60,80,100])
plt.yticks([0,50,100,150,200])
ax.set_xlim(-10, 110)
ax.set_ylim(-10, 210)
ax.set_xlabel('x')
ax.set_ylabel('y')
x =np.array([0,100])
y = 2*x
ax.plot(x,y,'b')
plt.show()
```

## Exercise 3

**Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig= plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
plt.xticks([0,0.2,0.4,0.6,0.8,1])
plt.yticks([0,0.2,0.4,0.6,0.8,1])
plt.show()
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

ax1.set_xticks([0,20,40,60,80,100])
ax1.set_yticks(np.array([0,2,4,6,8,10])*1000)

ax2.set_xticks([20,20.5,21,21.5,22])
ax2.set_yticks([30,35,40,45,50])

ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

ax1.set_xlim(0,100)
ax1.set_ylim(0,10000)

ax2.set_title('zoom')

ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2.set_xlabel('X')
ax2.set_ylabel('Y')

x =np.linspace(0,100,500)
z = x**2
y = (x**2)/10
ax1.plot(x,z,'b')
ax2.plot(x,y,'b')
plt.show()
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig = plt.figure()
fig,axes = plt.subplots(nrows=1, ncols=2)
plt.show()
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python

fig,axes = plt.subplots(nrows=1, ncols=2)
ax1 = axes[0]
ax2 = axes[1]
x = np.linspace(0,100,500)
z = x**2
y = x*2
ax1.set_xticks(np.linspace(0,100,5))
ax1.set_yticks(np.linspace(0,200,9))

ax2.set_xticks(np.linspace(0,100,5))
ax2.set_yticks(np.linspace(0,10000,6))

ax1.set_xlim(-10,110)
ax1.set_ylim(-10,210)

ax2.set_xlim(-10,110)
ax2.set_ylim(-1000,11000)

ax1.plot(x,y,'b--')
ax2.plot(x,z,'r')
plt.show()
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig,axes = plt.subplots(nrows=1, ncols=2,figsize=(15,3),dpi = 100)
ax1 = axes[0]
ax2 = axes[1]
x = np.linspace(0,100,500)
z = x**2
y = x*2
ax1.set_xticks(np.linspace(0,100,5))
ax1.set_yticks(np.linspace(0,200,9))

ax2.set_xticks(np.linspace(0,100,5))
ax2.set_yticks(np.linspace(0,10000,6))

ax1.set_xlim(-10,110)
ax1.set_ylim(-10,210)

ax2.set_xlim(-10,110)
ax2.set_ylim(-1000,11000)

ax1.plot(x,y,'b--',linewidth = 3)
ax2.plot(x,z,'r',linewidth = 1)
plt.show()
```
