```python
from numpy._core.defchararray import title
from pandas._libs import interval
import yfinance as yf
import plotly as pt
import pandas as pd
tickersymbol="AMZN"
ticker=yf.Ticker(tickersymbol)
hd=ticker.history(period='6mo',interval='1d')
print(hd)
print(hd.columns)
fig=pt.graph_objects.Figure(pt.graph_objects.Candlestick(x=hd.index,open=hd['Open'],close=hd['Close'],high=hd['High'],low=hd['Low'],))
fig.update_layout(title_text="AMAZON STOCK REPORT")
fig.show()
```
