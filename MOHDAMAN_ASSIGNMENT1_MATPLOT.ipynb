```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
fig=plt.figure(figsize=(100,6))
plt.subplot(1,2,1)
plt.plot(x,y,linestyle='dotted')
plt.subplot(1,2,2,)
plt.plot(x,z,marker='*')
plt.show()
```
