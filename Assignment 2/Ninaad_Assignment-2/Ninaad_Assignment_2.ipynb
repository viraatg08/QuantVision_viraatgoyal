```python
import yfinance as yf
import pandas as pd
```

```python
ticker = "MSFT"
data = yf.download(ticker, period = "6mo", interval = "1d")
data.reset_index(inplace=True)

data.to_csv('MSFT_Data.csv')
data = pd.read_csv('MSFT_Data.csv')
```

```python
import plotly.graph_objects as go
```

```python
candlestick = go.Candlestick(
    x=data['Date'],
    open=data['Open'],
    high=data['High'],
    low=data['Low'],
    close=data['Close']
)

fig = go.Figure(data=[candlestick])
```

```python
fig.update_layout(
    title=f"{ticker} - Last 6 Months",
    xaxis_title="Date",
    yaxis_title="Price"
)

fig.show()
```
