```python
# Importing Libraries
import yfinance as yf
import plotly.graph_objects as go
```

```python
ticker_symbol = "TSLA"
ticker = yf.Ticker(ticker_symbol)

# Downloading financial data
historical_data = ticker.history(period="6mo", interval="1d")
print(f"Historical Data {historical_data}")

# Converting Date from index to a proper column
historical_data = historical_data.reset_index()

# Plotting Candlestick chart
fig = go.Figure(data = go.Candlestick(
                                      x = historical_data['Date'],  # Mapping date column to the X axis slot
                                      open = historical_data['Open'],  # Mapping "Open" column to "open" parameter
                                      high = historical_data['High'],  # Mapping "High" column to "high" parameter
                                      low = historical_data['Low'],  # Mapping "Low" column to "low" parameter
                                      close = historical_data['Close']  # Mapping "Close" column to "close" parameter
                                      ))

# Adding Titles
fig.update_layout(
                  title='Tesla Stock Price Analysis (6 Months)',
                  yaxis_title='Stock Price (USD)',
                  xaxis_title='Date')

fig.show()




```

### Stock Pattern Analysis

* **Date:** 5th September, 2025
* **Pattern Name:** Inverted Hammer
* **Outcome:** The price went **UP** immediately after this pattern.
