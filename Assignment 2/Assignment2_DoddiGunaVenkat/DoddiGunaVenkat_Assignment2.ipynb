```python
import yfinance as yf
import plotly.graph_objects as go
```

## Step 2: Select a stock (any ticker of your choice) and use yfinance to download historical data for this stock. Get the last 6 months of data and set the interval to 1 day

```python
stock = "GOOGL" # selected google stock
ticker = yf.Ticker(stock)
stock_data = ticker.history(period="6mo",interval="1d") # set period of 6 months at 1day interval

stock_data.reset_index(inplace=True)
stock_data.head()
```

## Step 3: Now, visualise the numbers. Write code to create an interactive chart:
 - Use plotly.graph_objects to create a Candlestick chart.
 - Map the dataframe columns (Open, High, Low, Close) to the corresponding parameters in the chart function.
 - Add a title to the chart

**Requirement: The chart must be interactive (meaning you can zoom in and out to see individual candles clearly).**


```python
dates  = stock_data["Date"].values # dates
opens  = stock_data["Open"].values # open price
highs  = stock_data["High"].values # high price
lows   = stock_data["Low"].values # Low price
closes = stock_data["Close"].values # Closing price

fig = go.Figure(data=[go.Candlestick(x=dates, open=opens, high=highs, low=lows, close=closes)])
fig.update_layout(title="GOOGLE - Last 6 months Candlestick chart",
                  yaxis_title='Stock Price(US Dollar)',
                  xaxis_title='Date') # go interactive figure

fig.show()
```
